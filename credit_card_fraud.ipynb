```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# carregando o conjunto de dados para um Pandas DataFrame
credit_card_data = pd.read_csv("/creditcard.csv")
# primeiras 5 linhas do conjunto de dados
credit_card_data.head()
```

```python
# informaÃ§Ãµes do conjunto de dados
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
result = credit_card_data
```

```python
result = result.dropna()
```

```python
result.isnull().sum()
```

```python
result['Class'].value_counts()
```

This Dataset is highly unbalanced

0 â†’ Normal Transaction

1 â†’ Fraudulent transaction

```python
# separando os dados para anÃ¡lise
legit = result[result.Class == 0 ]
fraud = result[result.Class == 1 ]
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
result.groupby( 'Class' ).mean()
```

```python
legit_sample = legit.sample(n=88)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

Treinamento de modelo

empregamos RegressÃ£o LogÃ­stica, inicializando o modelo com parÃ¢metros padrÃ£o. Usando os dados de treinamento (X_train e Y_train), ajustamos o modelo de RegressÃ£o LogÃ­stica para aprender padrÃµes nos dados. Esse processo envolve o ajuste de parÃ¢metros do modelo para minimizar a diferenÃ§a entre rÃ³tulos de classe reais e previstos.

```python
model = LogisticRegression()
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

A pontuaÃ§Ã£o de precisÃ£o nos dados de treinamento Ã© de aproximadamente 98%. Isso indica que o modelo de RegressÃ£o LogÃ­stica prevÃª corretamente os rÃ³tulos de classe para cerca de 98% das instÃ¢ncias no conjunto de dados de treinamento.

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```
